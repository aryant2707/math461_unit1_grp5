## Long Term Actuarial Math Basics

An important concept in life insurance is the time value of money. Since money today is worth more than the same nominal amount in the future, we must discount future money to the present in order to compare them. We will use the force of interest $\delta$, which is the continuous compounding equivalent of an interest rate, in order to do this. To discount a cashflow of $D$ dollars $t$ years in the future, we calculate the present value as 
$$PV = D e^{-\delta \cdot t}$$

We model a person aged $x$'s future lifetime as a random variable $T_x$, who's value represents the number of years that a person lives beyond age $x$. The survival function this random variable, which represents The probability that a life aged $x$ lives beyond $t$ years is $P(T_x >t) = S_{T_x}(t) = {}_t p_x$, where ${}_t p_x$ is standard actuarial notation for the survival function.

Another important related function is the force of mortality $\mu_{x+t}$, also known as the hazard rate function for the random variable $T_x$. It roughly represents the probability density that a person dies at exactly time $x+t$. It can be written as follows
$$\mu_{x+t} = \dfrac{f_{T_x} (t)}{S_{T_x} (t)}$$

We can expand upon this to calculate the present value of life insurance and life annuities on a person aged $x$. For example, if a 40 year old person with a life insurance death benefit of $\$1000$ dies 20 years in the future at age 60, the present value of that future payment is 
$$1000 \cdot e^{-20 \delta}$$
where $\delta$ is the force of interest. We can write the present value of a life insurance of $\$1$ on a life aged $x$ in terms of the future lifetime random variable $T_x$ as $e^{-\delta \cdot T_x}$. We can derive a formula to calcul ate the expected present value of this life insurance.
$$\bar{A}_x = \int_{0}^{\infty} e^{-\delta t} \cdot {}_t p_x \cdot \mu_{x+t} \, dt$$
For every possible value $T_x = t$, $e^{-\delta t}$ represents the discounted present value of the payount, and ${}_t p_x \cdot \mu_{x+t}$ represents the probability that the life survives up to time $t$, then dies at exactly time $t$.

Similarily, we can calculate the expected present value of a life annuity of $\$1$ per year, payable continuously as
$$\bar{a}_x = \int_{0}^{\infty} e^{-\delta t} \cdot {}_t p_x   \, dt$$
$e^{-\delta t}$ represents the discounted present value of a cashflow paid at time $t$, and ${}_t p_x$ represent the probability that the policyholder survives to time $t$ to receive the payment.
